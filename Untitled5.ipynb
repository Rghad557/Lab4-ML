```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')

# pd.set_option("display.max_columns",None)
df = pd.read_csv("student_data.csv")

df.head()
```

```python
df['schoolsup'] = df['schoolsup'].map({'yes':1, 'no':0})
df['famsup'] = df['famsup'].map({'yes':1, 'no':0})
df['paid'] = df['paid'].map({'yes':1, 'no':0})
df['activities'] = df['activities'].map({'yes':1, 'no':0})
df['internet'] = df['internet'].map({'yes':1, 'no':0})
df['romantic'] = df['romantic'].map({'yes':1, 'no':0})
```

```python
df['sex'] = df['sex'].replace({'F':0, 'M':1})
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['G3'])

plt.title('Boxplot of G3')
plt.show()
```

```python
Q1 = df['G3'].quantile(0.25)
Q3 = df['G3'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
```

```python
outliers = df[(df['G3'] < lower) | (df['G3'] > upper)]
outliers.head()
```

```python
df_no_outliers = df[(df['G3'] >= lower) & (df['G3'] <= upper)]

print('df with outliers:', df.shape)
print('df without outliers:', df_no_outliers.shape)
```

```python
lower_cap = df['G3'].quantile(0.05)
upper_cap = df['G3'].quantile(0.95)

df_capped = df.copy()
df_capped['G3'] = df_capped['G3'].clip(lower_cap, upper_cap)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_scaled = df[['age','G1','G2','G3','absences']].copy()

df_scaled[['age','G1','G2','G3','absences']] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_standardized = df[['age','G1','G2','G3','absences']].copy()

df_standardized[['age','G1','G2','G3','absences']] = scaler.fit_transform(df_standardized)

df_standardized.head()
```

```python
sns.heatmap(df[['G1','G2','G3']].corr(), annot=True)
plt.title("Correlation between G1, G2, G3")
plt.show()
```

```python
from sklearn.decomposition import PCA

# PCA
X = df_standardized[['G1','G2','G3']]
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

# الرسم
plt.figure(figsize=(6,4))
plt.scatter(principal_components[:,0], principal_components[:,1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection")
plt.show()
```
